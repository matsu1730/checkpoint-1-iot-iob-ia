```python
# função para converter campos em one-hot

def one_hot(categorias, valor):
    categorias = list(set(categorias))
    indice = categorias.index(valor)
    vetor = np.zeros(len(categorias))
    vetor[indice] = 1
    return vetor
```

# normalização dos valores escalares

```python
def normaliza_media(valores):
    media = np.mean(valores)
    minimo = np.min(valores)
    maximo = np.max(valores)
    return (valores - media)/(maximo-minimo)
```

```python
df.degmalig = normaliza_media(df.degmalig)
df.head(10)
```

```python
# separando o vetor de classificação do resto do dataset
labels = df.classe.values
```

```python
#apagando coluna do conjunto de dados
df = df.drop(['classe'], axis=1)
df.head(10)
```

```python
# transformar conjunto de dados em vetores, com a função abaixo

def vetorizar(age, menopause, tumorsize, invnodes, degmalig, irradiant):
    return age.tolist() + menopause.tolist() + tumorsize.tolist() + invnodes.tolist() + [degmalig] + irradiant.tolist()
```

```python
entrada = [vetorizar(*_) for _ in df.values]
entrada
```

```python
# salvando dataset formatado para uso na Classificação e labels
np.savetxt("/content/entrada_breast_cancer.csv", entrada, delimiter=',')
np.savetxt("/content/labels.csv", labels.tolist(), delimiter=',')
```

# Classificação

```python
# importando o que é necessário
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
# carregando o dataset e labels
x = np.loadtxt("/content/entrada_breast_cancer.csv", delimiter=',')
y = np.loadtxt("/content/labels.csv", delimiter=',')
```

```python
x
```

```python
# criando modelo para classificação
model = KNeighborsClassifier(n_neighbors=5)
```

```python
# separando dados para treino e validação
tamanho_treino = int(x.shape[0]*0.7)
tamanho_treino
```

```python
x.shape[0] #linhas
```

```python
x.shape[1] #colunas
```

```python
# separando dados para treino
x_t = x[ :tamanho_treino]
y_t = y[ :tamanho_treino]
```

```python
# separanado para validação
x_v = x[tamanho_treino: ]
y_v = y[tamanho_treino: ]
```

```python
# criação do modelo
model.fit(x_t, y_t)
```

```python
# performance do modelo em dados de validação
model.score(x_v, y_v)
```
